```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
 x = np.linspace(0,10,11)
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x, x, label='X vs X', color='#e6c017', lw=2, ls='solid', marker = 'o', markersize=10) 
# Use RGB hex color; search hex color picker
# lw means line width; defauly value is 1.
# possible linetype options '-', '--', '-.', ':', 'None', ' ', '', 'solid', 'dashed', 'dashdot', 'dotted'

#ax.plot(x, x, label='X vs X', color='blue') # simplest way of coloring the plot lines or points

ax.plot(x, x **2, label= 'X vs X^2', color ='skyblue', linewidth=4, linestyle='-.')

#ax.legend()

#ax.legend(loc=4) # loc=0 is by default; this type of loc will put legend inside the graph

# To put legend outside the graph
ax.legend(loc=(1.1,0.5))
```
