```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(0,10)
```

```python
y = 2*x
```

```python
x
```

```python
y
```

```python
plt.plot(x,y)
plt.title('String Title')
plt.xlabel('X label')
plt.ylabel('Y label')
plt.xlim(0,9)
plt.ylim(0,18)
# semicolon make sure we only see rendering of visualization/same work as plt.show()
# If we don't use either ; or plt.show() then it will render the type of array and details about from which memory location it is picked
#write all code related to one plot in one jupyter cell else jupyter might through errors

plt.savefig('myfirstplot.png')
plt.show()
```

```python
#plt.figure() is used to create a figure object
fig1 = plt.figure() # fig is our first object
# now to see the plot we have to add axes to it
axes = fig1.add_axes([0,0,1,1]) # first two parameters are for x,y and last two are for width and height
axes.plot(x,y);
```

```python
a = np.linspace(0,10,11)
```

```python
b =  a**4
```

```python
b
```

```python
fig = plt.figure(figsize=(10,3), dpi=100)

# Large axes
axes1 = fig.add_axes([0,0,1,1])
axes1.plot(a,b)

axes1.set_xlim(0,10)
axes1.set_ylim(0,8000)

axes1.set_xlabel('A')
axes1.set_ylabel('B')

axes1.set_title('Power of 4')

# Small axes
#axes2 = fig.add_axes([1,1,0.3,0.3])
axes2 = fig.add_axes([0.2,0.2,0.3,0.3])
axes2.plot(a,b)


axes2.set_xlim(1,2)
axes2.set_ylim(0,50)

axes2.set_xlabel('A')
axes2.set_ylabel('B')

axes2.set_title('Zoomed in')


fig.savefig('axes_figure.png', bbox_inches = 'tight')

plt.show()
```

```python
fig_n, axes_n = plt.subplots(nrows=2, ncols=2, figsize = (10,8), dpi=200)

axes_n[0,0].plot(x,y)

axes_n[0,1].plot(a,b)
axes_n[0,1].set_title('Title 1,1')

fig_n.tight_layout() #removes any overlap between between two plots

fig_n.suptitle('Figure level', fontsize =16)

#fig_n.set_figwidth(10) #if not setting figure size

plt.show()

fig_n.savefig('newsubplot.png', bbox_inches = 'tight')

#fig_n.subplots_adjust(wspace=1, hspace=.9)
```
